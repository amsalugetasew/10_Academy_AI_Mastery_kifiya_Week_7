```python
import re
import emoji  # Don't forget to import the emoji library
```

```python
# Function to extract all URLs (any type of link)
def extract_links(text):
    # Regex to match any URL starting with http or https
    return re.findall(r'https?://[A-Za-z0-9./?&=_-]+', text)

# Function to extract emojis using a regex pattern
def extract_emojis(text):
    # Regex pattern for matching emojis
    emoji_pattern = re.compile(
        "[\U0001F600-\U0001F64F"  # Emoticons
        "\U0001F300-\U0001F5FF"  # Symbols & pictographs
        "\U0001F680-\U0001F6FF"  # Transport & map symbols
        "\U0001F700-\U0001F77F"  # Alchemical symbols
        "\U0001F780-\U0001F7FF"  # Geometric shapes
        "\U0001F800-\U0001F8FF"  # Supplemental arrows
        "\U0001F900-\U0001F9FF"  # Supplemental symbols and pictographs
        "\U0001FA00-\U0001FA6F"  # Chess symbols
        "\U0001FA70-\U0001FAFF"  # Symbols
        "\U00002702-\U000027B0"  # Dingbats
        "]+", flags=re.UNICODE)
    return ''.join(re.findall(emoji_pattern, text))

# Function to clean the message (remove links and emojis)
def clean_message(text):
    # Remove all links (http or https)
    text_no_links = re.sub(r'https?://[A-Za-z0-9./?&=_-]+', '', text)
    # Remove emojis
    text_no_emoji = re.sub(r'[\U0001F600-\U0001F64F'
                           r'\U0001F300-\U0001F5FF'
                           r'\U0001F680-\U0001F6FF'
                           r'\U0001F700-\U0001F77F'
                           r'\U0001F780-\U0001F7FF'
                           r'\U0001F800-\U0001F8FF'
                           r'\U0001F900-\U0001F9FF'
                           r'\U0001FA00-\U0001FA6F'
                           r'\U0001FA70-\U0001FAFF'
                           r'\U00002702-\U000027B0'
                           r']+', '', text_no_links)
    return text_no_emoji

# Apply functions to the DataFrame
df_pg['youtube_links'] = df_pg['message'].apply(extract_links)
df_pg['emoji_used'] = df_pg['message'].apply(extract_emojis)
df_pg['message'] = df_pg['message'].apply(clean_message)

# Output the DataFrame
df_pg[['message', 'emoji_used', 'youtube_links']]
```
